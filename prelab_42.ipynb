```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

print("OpenCV:", cv2.__version__)
print("NumPy:", np.__version__)
```

```python
# Step 3: Load Image
img = cv2.imread('CM23042.jpg')

if img is None:
    print("Image not found!")
else:
    print("Image loaded successfully.")

# Step 4: Display Original Image
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,6))
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# Step 5: Image Properties
print("Shape:", img.shape)
print("Height:", img.shape[0])
print("Width:", img.shape[1])
print("Channels:", img.shape[2])
print("Data Type:", img.dtype)
```

```python
# Step 6: Red Channel
plt.figure(figsize=(5,5))
plt.imshow(img_rgb[:,:,0], cmap='Reds')
plt.title("Red Channel")
plt.axis('off')
plt.show()

# Step 7: Green Channel
plt.figure(figsize=(5,5))
plt.imshow(img_rgb[:,:,1], cmap='Greens')
plt.title("Green Channel")
plt.axis('off')
plt.show()

# Step 8: Blue Channel
plt.figure(figsize=(5,5))
plt.imshow(img_rgb[:,:,2], cmap='Blues')
plt.title("Blue Channel")
plt.axis('off')
plt.show()
```

```python
# Step 9: Convert to Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(5,5))
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
# Step 11: Split Color Channels
r = img_rgb[:,:,0]
g = img_rgb[:,:,1]
b = img_rgb[:,:,2]

print("Red Channel Shape:", r.shape)
print("Green Channel Shape:", g.shape)
print("Blue Channel Shape:", b.shape)
```
